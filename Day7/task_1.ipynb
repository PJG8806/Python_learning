- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
num_list = []

while True:
    try:
        num_input = input("숫자를 입력해주세요(종료는 done입력)")
        if num_input == "done":
            break
        num_list.append(int(num_input))        
    except ValueError:
        print(f"잘못된 값: {num_input}. 숫자를 입력해주세요")
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time

def timer_decorator(delay):
    def timer():
        strt = time.time()
        delay()
        end = time.time()
        print(f"함수 실행 시간: {end-strt:.2f}초")
    return timer

@timer_decorator
def delay():
    time.sleep(0.1)

delay()
````

(예: 삼각형 넓이, 원의 넓이, 직육면체의 넓이)을 수행하는 함수들을 포함하는 mymath.py라는 이름의 모듈을 만드세요. 

이 모듈을 사용하여 작성된 함수들을 실제로 임포트하고 사용하는 간단한 프로그램을 작성하세요.

````python
import math

def triang(down_size,height_size):
    return down_size * height_size * 0.5

def circle(radius):
    return (radius ** 2) * math.pi

def cuboid(width1, whidth2, height):
    return ((width1 * height) + (whidth2 * height) + (width1 * whidth2)) * 2
````

````python
import mymath

print(mymath.triang(4,3))
print(mymath.circle(5))
print(mymath.cuboid(4,3,5))
````
